```python
import cobra.test

model = cobra.test.create_test_model("ecoli")
```

You can see a short summary of the loaded model with a few print commands:

```python
print len(model.reactions)
print len(model.metabolites)
print len(model.genes)
```

```python
model.reactions
```

We can see that the reactions are a list of object instances of Reaction class. We can use the cobrapy documentation to learn how to manipulate them, but in general cases when the documentation of a package would be poorly written or obsolete it is also possible to use the general mechanism of object introspection in Python. Sometimes it is convenient to look at the source code in order to understand how a Python library works when documentation is poor (it is not the case in this particular package).

```python
print type(model.reactions[0])
print dir(model.reactions[0])
```

So now we can see precisely what is the class of the Reaction object and we get a hint for where to find the source file. We also see what this object contains, what are its parameters and methods.

```python
print model.reactions[0].metabolites
print model.reactions[0].annotation
print model.reactions[0].reactants
print model.reactions[78].products
```

```python
for r in model.reactions:
    if len(r.products) > 0: print r, len(r.products), [m.name for m in r.products]
```

Above, I guessed the metabolite name attribute, but finding out the structure of a Metabolite object follows the same lines as in the Reaction case.
